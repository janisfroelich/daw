```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_stats = pd.DataFrame({"Jahr": range(1980,2021), "Eingeloeste_Autos": range(1980,2021), "Todesfälle": range(1980,2021), "Todeswahrscheinlichkeit_steinschlag": 0.0000713, "Todeswahrscheinlichkeit_Grenzwert":0.0001})
df_stats.set_index("Jahr", inplace=True)
```

```python
#read csv with multiple sheets
year = 1980
lst_tote = []
lst_year = []
while year <= 2020:
    df = pd.read_excel("je-d-11.06.01.01.01.xlsx", sheet_name=str(year), header=[2,3,4,5], index_col=0, nrows=32)
    df_stats["Todesfälle"][year] = df[("Verunfallte Personen","Alle Verunfallten", "Getötete", "Unnamed: 7_level_3")]["Total"]
    year += 1
```

```python
df_2010_2021 = pd.read_excel("cc-d-11.03.02.01.01.xlsx", sheet_name="2010-2021", header=1, index_col=1, nrows=10)
df_1990_2009 = pd.read_excel("cc-d-11.03.02.01.01.xlsx", sheet_name="1990-2009", header=1, index_col=1, nrows=10)
df_1980_1989 = pd.read_excel("cc-d-11.03.02.01.01.xlsx", sheet_name="1980-1989", header=1, index_col=1, nrows=10)
```

```python
df_1980_1989.dropna(axis=0, how="all", inplace=True)
```

```python
df_1980_1989.rename(columns={"1988 4)": 1988}, inplace=True)
```

```python
df_2010_2020 = df_2010_2021.drop(columns=[2021])
df_2010_2020
```

```python
df_2010_2021.drop(columns=["FAZG"], inplace=True)
df_1980_1989.drop(columns=["FAZG"], inplace=True)
df_1990_2009.drop(columns=["FAZG"], inplace=True)
```

```python
df_1980_1989.info()
```

```python
for i in df_2010_2020.columns:
    df_stats["Eingeloeste_Autos"][i] = df_2010_2020[i]["Total Fahrzeuge (ohne Mofas)"]
for i in df_1990_2009.columns:
    df_stats["Eingeloeste_Autos"][i] = df_1990_2009[i]["Total Fahrzeuge"]
for i in df_1980_1989.columns:
    df_stats["Eingeloeste_Autos"][i] = df_1980_1989[i]["Total Fahrzeuge"]
```

```python
df_stats["Todeswahrscheinlichkeit"] = df_stats["Todesfälle"] / df_stats["Eingeloeste_Autos"]
df_stats
```

```python
#lineplot with seaborn library and probability of death
plt.subplots(figsize=(10,5))
sns.lineplot(x="Jahr", y="Todeswahrscheinlichkeit_Grenzwert", data=df_stats, label="Grenzwert", linewidth=2)
plt.title("Interpretation Grenzwert Todesfallwahrscheinlichkeit", fontsize=12)
sns.set(font_scale=1)
plt.ylim(0, 0.0005)
plt.show()
```

```python
#lineplot with seaborn library and probability of death
plt.subplots(figsize=(10,5))
sns.lineplot(x="Jahr", y="Todeswahrscheinlichkeit", data=df_stats, label="Strassenverkehr", linewidth=2)
sns.lineplot(x="Jahr", y="Todeswahrscheinlichkeit_steinschlag", data=df_stats, label="Kantonstrasse", linewidth=2)
sns.lineplot(x="Jahr", y="Todeswahrscheinlichkeit_Grenzwert", data=df_stats, label="Grenzwert", linewidth=2)
plt.title("Vergleich Todesfallwahrscheinlichkeit Strassenverkehr über Zeit mit Todesfallwahrscheinlichkeit Kantonsstrasse", fontsize=12)
sns.set(font_scale=1)
plt.ylim(0, 0.0005)
plt.show()
```
